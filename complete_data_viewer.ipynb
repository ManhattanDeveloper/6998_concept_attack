```python
import pandas
import sqlite3
```

```python
conn = sqlite3.connect("complete_data.db")
cur = conn.cursor()
```

```python
pandas.read_sql_query("select * "+
                      "from complete_table "+
                      "limit 2 "+
                      ";", conn)
```

```python
pandas.read_sql_query("select dataset_name, label, avg(score) "+
                      "from complete_table "+
                      "where label='airplane'"+
                      "group by dataset_name, label"
                      ";", conn)
```

```python
pandas.read_sql_query("select * from (select dataset_name, category, count(*) "+
                      "from complete_table "+
                      "group by dataset_name, category "
                      ") groups, complete_table;", conn)
```

```python
pandas.read_sql_query("select unit, count(distinct label) as counts "+
                      "from complete_table "+
                      "group by unit "+
#                       "limit 10 "+
                      "order by counts DESC"
                      ";", conn)
```

```python
pandas.read_sql_query("select count(*) "+
                      "from complete_table "+
                      "group by unit "+
#                       "limit 10 "+
#                       "order by counts DESC"
                      ";", conn)
```

```python
pandas.read_sql_query("select b.dataset_name, count(*) "+
                      "from complete_table a, complete_table b "+
                      "where a.unit=b.unit and a.label!=b.label and a.dataset_name='original_df' "
                      "group by b.dataset_name "+
                      ";", conn)
# .to_csv('changed_unit_category_by_dataset_by_category.csv')
```
